# Post-Run Analysis

Load pipeline results and generate analytical visualizations:
- Detection count distribution
- Per-track trajectory lengths
- Model comparison summary

```python
import sys
sys.path.insert(0, "..")

import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

%matplotlib inline
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)

# Load pipeline run summary
summary_path = "../data/output/run_summary.json"
with open(summary_path) as f:
    summary = json.load(f)

print("Pipeline Run Summary:")
for k, v in summary.items():
    print(f"  {k}: {v}")
```

## Model Comparison Results

```python
# Load model comparison data
comp_path = "../report/figures/model_comparison.json"
if Path(comp_path).exists():
    with open(comp_path) as f:
        comp_data = json.load(f)
    
    print(f"{'Model':<15} {'Tracker':<12} {'FPS':>8} {'Unique IDs':>12} {'Avg Det/Frame':>15}")
    print("-" * 65)
    for r in comp_data:
        print(f"{r['model']:<15} {r['tracker']:<12} {r['fps']:>8.1f} {r['unique_ids']:>12} {r['avg_detections_per_frame']:>15.1f}")
    
    # Display the comparison chart
    from IPython.display import Image
    chart_path = "../report/figures/model_comparison.png"
    if Path(chart_path).exists():
        display(Image(filename=chart_path))
else:
    print("No model comparison data found. Run the pipeline_demo notebook first.")
```

## View Generated Assets

```python
from IPython.display import Image
import glob

# Display all screenshots
screenshots = sorted(glob.glob("../data/screenshots/*.png"))
print(f"Found {len(screenshots)} screenshots")

if screenshots:
    fig, axes = plt.subplots(1, min(len(screenshots), 5), figsize=(20, 5))
    if not hasattr(axes, '__len__'):
        axes = [axes]
    for ax, path in zip(axes, screenshots[:5]):
        import cv2
        img = cv2.imread(path)
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        ax.set_title(Path(path).stem, fontsize=9)
        ax.axis("off")
    plt.suptitle("Sample Screenshots from Pipeline Run", fontsize=14)
    plt.tight_layout()
    plt.show()

# Display heatmap if available
heatmap_path = "../report/figures/movement_heatmap.png"
if Path(heatmap_path).exists():
    print("\nMovement Heatmap:")
    display(Image(filename=heatmap_path))

# Display object count chart
count_path = "../report/figures/object_count.png"
if Path(count_path).exists():
    print("\nObject Count Over Time:")
    display(Image(filename=count_path))
```
